```python
import pandas as pd
import numpy as np
```

```python
# Loading CSV file using latin1 encoding
# because the dataset is not encoded in UTF-8

df = pd.read_csv(
    r"C:\Users\geeta\OneDrive\Desktop\coding\Python\Projects\Business KPI Dashboard\Data\raw_data.csv",
    encoding="latin1"
)

print(df.head())
```

```python
df.duplicated().sum()

df = df.drop_duplicates()
```

```python
# Convert Date Columns
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Profit Margin %'] = (df['Profit'] / df['Sales']) * 100
```

```python
df['Shipping Days'] = (
    df['Ship Date'] - df['Order Date']
).dt.days
```

```python
df['Month-Year'] = df['Order Date'].dt.to_period('M')
```

```python
loss_orders = df[df['Profit'] < 0]
```

```python
shipping_analysis = (
    df.groupby('Ship Mode')['Shipping Days']
    .mean()
    .sort_values()
)

shipping_analysis
```
